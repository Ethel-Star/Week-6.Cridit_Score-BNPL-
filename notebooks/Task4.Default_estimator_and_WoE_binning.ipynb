#### Default estimator and WoE binning 

The steps in building a credit scoring system using RFMS and WoE binning are essential for transforming raw transaction data into actionable insights. RFMS calculation helps quantify user behavior across recency, frequency, monetary value, and stability, which are key indicators of default risk. Classifying users into good or bad based on their RFMS score allows us to categorize users as high-risk or low-risk. 

WoE binning then converts continuous features into discrete bins, improving model interpretability and predictive power. Finally, feature engineering ensures that the data is structured to capture meaningful patterns, enhancing the accuracy and reliability of the model for predicting defaults.

In this task I used a featured engineered data because it helps turn raw data into meaningful inputs for machine learning models. In this case, it ensures the model can capture the most important aspects of customer behavior (like transaction frequency and value) and handle complex, non-linear relationships between those features and the likelihood of default. This step is vital for improving the accuracy and reliability of credit scoring models.
